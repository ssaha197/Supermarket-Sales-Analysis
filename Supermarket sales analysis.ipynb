### Gender and Customer type

```python
plt.figure(figsize=(12,5))
ax = sns.countplot('Gender', data = data, palette = 'mako')
ax.set_xlabel(xlabel = 'Gender')
ax.set_ylabel(ylabel = 'Gender Count')
ax.set_title(label = 'Gender Count in Supermarket', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(12,5))
ax = sns.countplot('Customer type', hue = 'Branch', data = data, palette = 'mako')
ax.set_xlabel(xlabel = 'Branches')
ax.set_ylabel(ylabel = 'Customer Count')
ax.set_title(label = 'Type of Customers in different branch', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(12,5))
ax = sns.countplot('Payment', hue = 'Branch', data = data, palette = 'mako')
ax.set_xlabel(xlabel = 'Payment method')
ax.set_ylabel(ylabel = 'Customer Count')
ax.set_title(label = 'Payment Distribution in all branches', fontsize = 20)
plt.show()
```

### Rating in different Branches

```python
plt.figure(figsize=(12,5))
ax = sns.boxplot(x = 'Branch', y = 'Rating', data = data, palette = 'rocket')
ax.set_xlabel(xlabel = 'Branches')
ax.set_ylabel(ylabel = 'Rating distribution')
ax.set_title('Rating distribution between branches')
```

### sales by time

```python
plt.figure(figsize=(15,5))
data['Time'] = pd.to_datetime(data['Time'])
data['Hour'] = (data['Time']).dt.hour
SalesTime = sns.lineplot(x = 'Hour', y = 'Quantity', data = data)
SalesTime.set_title('Product Sales per Hour')
```

### Rating vs Sales

```python
plt.figure(figsize=(15,5))
rvs = sns.lineplot(x = 'Total', y = 'Rating', data = data)
rvs.set_title('Rating vs Sales')
```

### Product line

```python
plt.figure(figsize=(12,5))
ax = sns.boxenplot(x = 'Quantity', y = 'Product line', data = data, palette = 'rocket')
ax.set_xlabel(xlabel = 'Qunatity Sales')
ax.set_ylabel(ylabel = 'Product Line')
ax.set_title(label = 'Average sales of different lines of products', fontsize = 20)
```

```python
plt.figure(figsize=(12,5))
ax = sns.countplot(y = 'Product line', data = data, order = data['Product line'].value_counts().index, palette = 'rocket')
ax.set_xlabel(xlabel = 'Sales count')
ax.set_ylabel(ylabel = 'Product Line')
ax.set_title(label = 'Sales count of products', fontsize = 20)
```

```python
plt.figure(figsize=(12,5))
ax = sns.boxenplot(x = 'Total', y = 'Product line', data = data, palette = 'rocket')
ax.set_xlabel(xlabel = 'Total Sales')
ax.set_ylabel(ylabel = 'Product Line')
ax.set_title(label = 'Total Sales of products', fontsize = 20)
```

### Average Ratings

```python
plt.figure(figsize=(12,5))
ax = sns.boxenplot(x = 'Rating', y = 'Product line', data = data, palette = 'rocket')
ax.set_xlabel(xlabel = 'Rating')
ax.set_ylabel(ylabel = 'Product Line')
ax.set_title(label = 'Average Rating of Product lines', fontsize = 20)
```
